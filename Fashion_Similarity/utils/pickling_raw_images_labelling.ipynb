```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import cv2
```

```python
def img_to_dataframe(row):
    try:
        path_to_image="../datasets/images/"+row
        img_array=cv2.imread(path_to_image)
        img_array=cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB ) 
        return img_array
    except:
        print("Skipping file due to Error - ../datasets/images/"+row)
        

        
df=pd.read_csv("../datasets/styles.csv", error_bad_lines=False, warn_bad_lines=False)
df=df.dropna().reset_index(drop=True)
df["image_name"]=df.apply(lambda row: str(row["id"])+".jpg",axis=1).reset_index(drop=True)
display(df)


df["image_array"]=df["image_name"].apply(img_to_dataframe)
df=df.dropna().reset_index(drop=True)
df.shape,display(df)
```

```python
df.to_csv("../datasets/Dataset_With_Image_Array.csv")
```
